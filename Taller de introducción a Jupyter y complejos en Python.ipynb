## Exercise 5: Complex Number Properties
1. Prove that the absolute value of a product of two complex numbers is the product of their absolute values.
2. Show that the conjugate of a sum of two complex numbers is the sum of their conjugates.

```python
import numpy as np

print("Exercise 5: Complex Number Properties")

# 1. Absolute value of a product
def abs_product(z1, z2):
    product = z1 * z2
    return abs(product)

def product_abs(z1, z2):
    return abs(z1) * abs(z2)

print("1. Absolute value of a product")
z1 = 3 + 4j
z2 = 1 - 2j
print(f"z1: {z1}")
print(f"z2: {z2}")
print(f"Absolute value of the product: |{z1} * {z2}| = {abs_product(z1, z2)}")
print(f"Product of the absolute values: |{z1}| * |{z2}| = {product_abs(z1, z2)}")
print("TRUE" if abs_product(z1, z2) == product_abs(z1, z2) else "FALSE")

z1 = -1 + 1j
z2 = 2 + 3j
print(f"\nz1: {z1}")
print(f"z2: {z2}")
print(f"Absolute value of the product: |{z1} * {z2}| = {abs_product(z1, z2)}")
print(f"Product of the absolute values: |{z1}| * |{z2}| = {product_abs(z1, z2)}")
print("TRUE" if abs_product(z1, z2) == product_abs(z1, z2) else "FALSE")

# 2. Conjugate of a sum
def sum_conjugate(z1, z2):
    return (z1 + z2).conjugate()

def conjugate_sum(z1, z2):
    return z1.conjugate() + z2.conjugate()

print("\n2. Conjugate of a sum")
z1 = 1 + 2j
z2 = 3 - 4j
print(f"z1: {z1}")
print(f"z2: {z2}")
print(f"Conjugate of the sum: ({z1} + {z2}).conjugate() = {sum_conjugate(z1, z2)}")
print(f"Sum of the conjugates: {z1}.conjugate() + {z2}.conjugate() = {conjugate_sum(z1, z2)}")
print("TRUE" if sum_conjugate(z1, z2) == conjugate_sum(z1, z2) else "FALSE")

z1 = -1 - 1j
z2 = 2 + 2j
print(f"\nz1: {z1}")
print(f"z2: {z2}")
print(f"Conjugate of the sum: ({z1} + {z2}).conjugate() = {sum_conjugate(z1, z2)}")
print(f"Sum of the conjugates: {z1}.conjugate() + {z2}.conjugate() = {conjugate_sum(z1, z2)}")
print("TRUE" if sum_conjugate(z1, z2) == conjugate_sum(z1, z2) else "FALSE")
```
